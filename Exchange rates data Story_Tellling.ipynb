Exploratory data visualization: we create graphs for ourselves to better understand and

explore data.

Explanatory data visualization: we create graphs for others to inform, make a point, or tell a story.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar','Period\\Unit:': 'Time'},inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time',inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != "-"]
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## rolling  mean


```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.show()
```

we look at the line's shape, we see many small wiggles - rather than seeing a smooth line. The wiggles, however, have meaning:: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

```python
values=pd.DataFrame()
values['daily_values']=pd.Series(range(1,20,2))
values
```

```python
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
```

```python
values
```

```python
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()
values
```

```python
plt.figure(figsize=(9,6))
```

```python
plt.figure(figsize=(9,6))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('Orignal values',weight='bold')
plt.ylim(0.8,1.7)
for i ,rolling_mean in zip([2,3,4,5,6],
                          [7,30,50,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
            euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean),weight='bold')

    plt.tight_layout()
    plt.show()
```

## story telling

```python
financial_crisis=euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2006)&(euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_=euro_to_dollar.copy()[(euro_to_dollar.Time.dt.year  >=2007)&(euro_to_dollar.Time.dt.year <=2008)]
```

```python
financial_crisis
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'financial_crisis' and 'financial_crisis_7_8' are DataFrames containing necessary data

# Setting the style
plt.style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=1, color='#A60785')

# Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'], linewidth=3, color='#e23d28')

# Setting x-axis tick labels
##ax.set_xticklabels(["2006", "", "2007", "", "2008", "", "2009", "", "2010"], alpha=0.6)

# Adding text annotations for years
x = 0.02
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=ax.transAxes)
    x += 0.2

# Setting y-axis tick labels
ax.set_yticklabels(["", "1.2", "1.3", "1.4", "1.5"], alpha=0.5)

# Adding text annotations for rates
y = 0.07
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(-0.04, y, rate, alpha=0.5, fontsize=11, transform=ax.transAxes)
    y += 0.2333

# Adding a title and a subtitle
ax.text(-0.05, 1.2, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight="bold", transform=ax.transAxes)
ax.text(-0.05, 1.1, 'Euro-USD exchange rates between 2006 and 2010', size=12, transform=ax.transAxes)

# Adding a signature
ax.text(-0.05, -0.25, '@TECHMA ZONE'+''*80+ "SOURCE: European Central Bank", color="#fefefe", fontsize=12, backgroundcolor="#4d4d4d", transform=ax.transAxes)
 
# Adding a shaded area for the crisis period
ax.axvspan(xmin=pd.to_datetime("2008-04-01"), xmax=pd.to_datetime("2008-09-01"), ymin=0.09, alpha=0.3, color="grey")

plt.show()

```

```python
corona_crisis_20 = euro_to_dollar.loc[(euro_to_dollar["Time"] >= '2020-01-01') &
                                       (euro_to_dollar["Time"] <= '2020-12-31')]

corona_crisis = euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2016-01-01') &
                                    (euro_to_dollar['Time'] <= '2019-12-31')]
```

```python
corona_crisis['rolling_mean'].value_counts(bins = 5).sort_index()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
import pandas as pd

# Assuming 'corona_crisis' and 'corona_crisis_20' are already defined DataFrames

style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize=(8, 3))

ax.plot(corona_crisis['Time'], corona_crisis['rolling_mean'],
        linewidth=1.5, color='#A6D785')
ax.plot(corona_crisis_20['Time'], corona_crisis_20['rolling_mean'],
        linewidth=3, color='#e23d28')

ax.grid(True, color='grey', linestyle='-', linewidth=0.5)

ax.tick_params(axis='x', which='both', colors='grey')
ax.tick_params(axis='y', which='both', colors='grey')

x = 0.02
for year in ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=plt.gca().transAxes)
    x += 0.183

ax.text(-0.05, 1.3, "Euro-USD rate peaked at 1.236 during 2020's Global Crisis",
        weight='bold', transform=plt.gca().transAxes)
ax.text(-0.05, 1.2, "Euro-USD exchange rates between 2016 and 2021",
        fontsize=12, fontweight='bold', transform=plt.gca().transAxes)
ax.text(-0.05, -0.25, "@TECHMA ZONE" + " " * 50 + "Source: European Central Bank",
        weight='bold', color='#f0f0f0', backgroundcolor='#4d4d4d', transform=plt.gca().transAxes)

ax.axvspan(xmin=pd.to_datetime("2021-01-21"), xmax=pd.to_datetime("2021-01-21"), alpha=0.3, color='grey')

plt.tight_layout()
plt.show()
```

## The Three Us _Presidents Example

```python
bush_obama_trump= euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2001)&(euro_to_dollar['Time'].dt.year <2021)]

bush=bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year <2009)]

obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2009)&(bush_obama_trump['Time'].dt.year < 2017)]

trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2017)&(bush_obama_trump['Time'].dt.year <2021)]
```

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.figure(figsize=(20, 15))

ax1 = plt.subplot(3, 3, 1)
ax2 = plt.subplot(3, 3, 2)
ax3 = plt.subplot(3, 3, 3)
ax4 = plt.subplot(3, 1, 2)

axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.6)

ax1.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3, size=12)
ax1.text(0.11, 2.55, "BUSH ERA", fontsize=20, weight="bold", color='#BF5FFF', transform=ax1.transAxes)
ax1.text(0.093, 2.35, "{2001-2009}", fontsize=20, weight="bold", color='grey', transform=ax1.transAxes)

ax2.plot(obama['Time'], obama['rolling_mean'], color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3, size=12)
ax2.text(0.11, 2.55, "OBAMA ERA", fontsize=20, weight="bold", alpha=0.3, color='#00B2EE', transform=ax2.transAxes)
ax2.text(0.093, 2.35, "{2009-2017}", fontsize=13, weight="bold", alpha=0.3, color='BLACK', transform=ax2.transAxes)

ax3.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE')
ax3.set_xticklabels(['', '2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3, size=12)
ax3.text(0.11, 2.55, "TRUMP ERA", fontsize=20, weight="bold", alpha=0.3, color='#00B2EE', transform=ax3.transAxes)
ax3.text(0.093, 2.35, "{2017-2021}", fontsize=13, weight="bold", alpha=0.3, color='BLACK', transform=ax3.transAxes)

ax4.plot(bush['Time'], bush['rolling_mean'], color='#BF5FFF', label='George W. Bush (2001-2009)')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#ffa500', label='Barack Obama (2009-2017)')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#00B2EE', label='Donald Trump (2017-2021)')

ax4.set_xticks([])
ax4.legend()

plt.figtext(0.5, 1.06, "EURO_USD rate averaged at 1.22 under the last three US presidents", fontsize=20, weight='bold', ha='center')
plt.figtext(0.5, 1.06, "EURO_USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)",
            fontsize=20, weight='bold',ha='center', transform=plt.gca().transAxes)

ax4.text(-0.005, -0.225, "@TECHMA ZONE" + " " * 105 + ' Source European Central Bank',
         color='#f0f0f0', backgroundcolor='#4d4d4d', fontsize=18, transform=ax4.transAxes)

plt.tight_layout()
plt.show()
```
